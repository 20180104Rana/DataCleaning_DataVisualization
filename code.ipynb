# Convert Data Type

```python
df['Tutoring_Sessions'] = df['Tutoring_Sessions'].astype("object")
df['Physical_Activity'] = df['Physical_Activity'].astype("object")
df['Sleep_Hours'] = df['Sleep_Hours'].astype("object")
```

# Repalce the null value from column

```python
df.isnull().sum()
```

```python
obj_col = df.select_dtypes(include=['object']).columns
num_col = df.select_dtypes(include=['number']).columns
```

```python
print(obj_col)
print(num_col)
```

```python
# Handle null values in objects
for column in obj_col:
  if df[column].isnull().any():
    random_values = df[column].dropna().sample(n=df[column].isnull().sum(), replace=True)
    df.loc[df[column].isnull(), column] = random_values.values
```

```python
# Handle null values in numbers
for column in num_col:
  if df[column].isnull().any():
    df[column].fillna(df[column].median(), inplace=True)
```

```python
def box_plot(column):
  plt.figure(figsize=(10, 6))
  sns.boxplot(x=column, data=df)
  plt.title(f'Box Plot of {column}')
  plt.show()
```

```python
for column in df.columns:
  if df[column].dtype != 'object':
    box_plot(column)
```

```python
# IQR : Interquantile range
def has_outliers(column):

  Q1 = column.quantile(0.25)
  Q3 = column.quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  return any((column < lower_bound) | (column > upper_bound))
```

```python
for column in df.columns:
  if df[column].dtype != 'object':
    if has_outliers(df[column]):
      print(f"Column '{column}' contains outliers.")
    else:
      print(f"Column '{column}' does not contain outliers.")
```

```python
def assign_grade(score):
  if score >= 90:
    return 'A'
  elif score >= 80:
    return 'B'
  elif score >= 70:
    return 'C'
  elif score >= 60:
    return 'D'
  else:
    return 'F'
```

```python
df['Grade'] = df['Exam_Score'].apply(assign_grade)

plt.figure(figsize=(10, 6))
sns.countplot(x='Grade', data=df, order=['A', 'B', 'C', 'D', 'F'], palette='viridis')
plt.xlabel('Grade')
plt.ylabel('Number of Students')
plt.title('Grades')
plt.show()
```
